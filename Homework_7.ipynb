```python
import numpy as np
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.stop_words import ENGLISH_STOP_WORDS
```

```python
newsgroups_train = fetch_20newsgroups(subset='train', remove=('headers', 'footers', 'quotes'))
```

Выведем список категорий текстов:

```python
newsgroups_train.target_names
```

```python
vectorizer = CountVectorizer(lowercase=True, stop_words=ENGLISH_STOP_WORDS,
                             analyzer='word', max_df = 0.11, min_df = 0.014, binary=True)
vectorizer.fit(newsgroups_train.data)
```

```python
len(vectorizer.vocabulary_)
```

```python
vectorizer.vocabulary_
```

```python
X_train = vectorizer.fit_transform(newsgroups_train.data)
X_train.shape
```

```python
def LDA(X, niter = 100, alpha = np.ones(20)):
    Numb = X.shape[1]
    temp = 20
    Betta = np.ones(Numb)
    Words = X.nonzero()[1]
    N_KW =  np.zeros((temp, Numb))
    Docs = X.nonzero()[0]
    N_DK = np.zeros((len(Docs), temp))
    nK = np.zeros(temp)
    tegs = np.random.choice(temp, len(Docs))
    for Doc, Word, teg in zip(Docs, Words, tegs):
        N_DK[Doc, teg] += 1
        N_KW[teg, Word] += 1
        nK[teg] += 1
    for i in range(niter):
        for j in range( len(Docs) ):
            teg = tegs[j]
            Doc = Docs[j]
            Word = Words[j]
            
            N_KW[teg, Word] -= 1
            N_DK[Doc, teg] -= 1
            nK[teg] -= 1
            
            P = np.array((N_DK[Doc, :] + alpha)*(N_KW[:, Word] + Betta[Word])/(nK + Betta.sum()))
            P /= P.sum()
            tegs[j] = np.random.choice(np.arange(temp), p = P)
            teg = tegs[j]
            
            N_KW[teg, Word] += 1
            N_DK[Doc, teg] += 1
            nK[teg] += 1

    T = np.array([np.argsort(N_KW[k, :]) for k in range(temp)])
    for k in range(temp):
        arr = np.zeros(Numb)
        for j in T[k, -11: -1]:
            arr[j] = 1
        print('Tag {}: {}'.format(k, '\\t'.join(str(x) for x in vectorizer.inverse_transform(arr))))
```

```python
LDA(X_train)
```

Вывод: слова хорошо разбросаны по тегам и хорошо прослеживаются темы. Teg 18 относится к религии , Tag 16 относится к политике и так далее.
