```python
import matplotlib.pyplot as plt
import numpy as np
```

# Task 1

Plot the populations for the years in the period. Add legend and axis labels to the plot.

```python
data = data.astype(int)
years = data[:,0]
hare = data[:,1]
lynx = data[:,2]
carrot = data[:,3]
```

```python
fig, ax = plt.subplots()

ax.plot(years, hare, label='Hare', c='r', linewidth=3)
ax.plot(years, lynx, label='Lynx', c='b', linewidth=3)

fig.set_figheight(6)
fig.set_figwidth(12)

plt.xticks(range(min(years), max(years) + 1))
plt.xlabel('Year')
plt.ylabel('Population')

plt.legend()
plt.show() 
```

Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each specis for the years in the period.

```python
plt.bar(years, hare / (hare+lynx), color = 'r')
plt.bar(years, lynx / (hare+lynx),  bottom = hare/ (hare+lynx),  color = 'b')
plt.ylabel('зайцы - красным/ лисы - синим')
plt.xlabel('years')
plt.show()
```

Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Arrange 3 subplots in a row.

Find which year each species had the largest population. Mark these years in the plot of populations.

```python
max1 = hare == np.max(hare)
max2 = lynx == np.max(lynx)
fig, ax = plt.subplots()
ax.plot(years, hare, label='Hare', c='lightgreen', linewidth=2)
ax.plot(years, lynx, label='Lynx', c='lightblue', linewidth=2)
ax.scatter(years[max1], hare[max1], color='g')
ax.scatter(years[max2], lynx[max2], color='b')
fig.set_figwidth(15)    
fig.set_figheight(6)
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend()
plt.show() 
```

Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing ratio of "H" and "L" values obtained.

```python
list = []
for i in range(len(years)):
    if (hare[i] > lynx[i]):
        list.append('H')
    else:
        list.append('L')
print(list[:])
```

```python
numberL = np.char.count(list, 'L')
numberH = np.char.count(list, 'H')

vals = [sum(numberL),sum(numberH)]
labels = ["Lynx", "Hare"]
fig, ax = plt.subplots()
ax.pie(vals, labels=labels)
ax.axis("equal")
plt.legend()
fig.set_figwidth(5)    
fig.set_figheight(5)
```

Plot the change in hare population and the number of lynxes. Find the correlation coefficient.

```python
HareDif = np.insert(np.diff(hare), 0, 0)
LynxDif = np.insert(np.diff(lynx), 0, 0)
fig, ax = plt.subplots()
ax.plot(years, HareDif, label='Change in hare population', c='y', linewidth=2)
ax.plot(years, LynxDif, label='Change in lynx population', c='blue', linewidth=2)

plt.xticks(range(min(years), max(years) + 1))
plt.xlabel('Year')
plt.ylabel('Population')

fig.set_figwidth(15)    
fig.set_figheight(5)

plt.legend()
plt.show() 

print("Correlation: ", np.corrcoef(HareDif, LynxDif)[0, 1])
```

Plot a graph of total explained variance vs number of principal components.

```python
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
digits = fetch_openml('mnist_784')
images = digits.data.reshape(-1, 28, 28)
```

```python
plt.figure(figsize = (20,8))
pca = PCA(n_components=784)
pca.fit_transform(digits.data.astype(float))
cumulative=[]
last=0
X_transformed = pca.fit_transform(digits.data.astype(float))
X_transformed.shape
for i in pca.explained_variance_ratio_:
    cumulative.append(last+i)
    last = cumulative[-1]
cumulative = np.asarray(cumulative)    
plt.plot(100*cumulative, c = 'black')
plt.xlabel('number of components')
plt.ylabel('total explained variance')
plt.show()
```
